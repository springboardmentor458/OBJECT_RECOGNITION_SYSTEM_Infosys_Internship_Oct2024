```python
import cv2
import json
from matplotlib import pyplot as plt

with open('key.json', 'r') as f:
    bounding_boxes = json.load(f)
image_path = '000000001503.jpg'  
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f"Image not found at {image_path}")
for box in bounding_boxes['annotations']:
    bbox = box['bbox']
    x, y, w, h = bbox
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title("Image with Bounding Boxes")
plt.show()
```

```python
import cv2
import json
from matplotlib import pyplot as plt
with open('key.json', 'r') as f:
    bounding_boxes = json.load(f)

image_path = '000000001503.jpg' 
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f"Image not found at {image_path}")

# Convert the image to grayscale for edge detection
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray_img, threshold1=100, threshold2=200)
edges_rgb = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(edges_rgb)
plt.axis('off')
plt.title("Edge Detection (Outline)")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image_path = '000000001503.jpg'  
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f"Image not found at {image_path}")
flipped_horizontal = cv2.flip(img, 1)  
flipped_vertical = cv2.flip(img, 0)    
flipped_both = cv2.flip(img, -1)       

flipped_horizontal_rgb = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flipped_vertical_rgb = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)
flipped_both_rgb = cv2.cvtColor(flipped_both, cv2.COLOR_BGR2RGB)

# Display the flipped images
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(flipped_horizontal_rgb)
plt.axis('off')
plt.title("Flipped Horizontally")

plt.subplot(1, 3, 2)
plt.imshow(flipped_vertical_rgb)
plt.axis('off')
plt.title("Flipped Vertically")

plt.subplot(1, 3, 3)
plt.imshow(flipped_both_rgb)
plt.axis('off')
plt.title("Flipped Both Directions")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
def rotate_image(image, angle):
    rotation_map = {
        90: cv2.ROTATE_90_CLOCKWISE,
        180: cv2.ROTATE_180,
        270: cv2.ROTATE_90_COUNTERCLOCKWISE
    }
    return cv2.rotate(image, rotation_map[angle])

# Load the image
image_path = '000000001503.jpg'  
img = cv2.imread(image_path)

if img is None:
    raise FileNotFoundError(f"Image not found at {image_path}")
angles = [90, 180, 270]
plt.figure(figsize=(12, 4))

# Loop to rotate
for i, angle in enumerate(angles):
    rotated_image = rotate_image(img, angle)
    rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)  # Color 
    
    # Display the rotated image
    plt.subplot(1, 3, i+1)
    plt.imshow(rotated_image_rgb)
    plt.title(f'Rotated {angle}')
    plt.axis('off')  

plt.tight_layout() 
plt.show()
```
