```python
import cv2
import matplotlib.pyplot as plt

def display_image_with_bboxes(image_path, bounding_boxes):
    # Load the image using OpenCV
    image = cv2.imread(image_path)
    
    # Convert the image to RGB (OpenCV loads images in BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Loop through the bounding boxes and draw them on the image
    for box in bounding_boxes:
        x, y, w, h = box
        # Draw a rectangle for each bounding box (x, y) is the top-left corner
        # (x+w, y+h) is the bottom-right corner
        cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    # Display the image with bounding boxes
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis
    plt.show()

# Example: Path to the image and sample bounding boxes
image_path = r"C:\Users\harsh\Downloads\val2017-20241106T125252Z-001\val2017\000000415194.jpg"  # Replace with the actual image path
# Bounding boxes in the format [x, y, w, h] where:
# x, y = top-left corner, w = width, h = height
bounding_boxes = [[50, 30, 100, 150], [200, 100, 120, 160]]  # Replace with actual coordinates

# Display the image with bounding boxes
display_image_with_bboxes(image_path, bounding_boxes)
```
