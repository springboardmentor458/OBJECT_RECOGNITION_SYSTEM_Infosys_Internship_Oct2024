```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/Valley.jpg')  # Replace with the path to your image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for correct color display

# Create a figure with a 1x2 layout
plt.figure(figsize=(12, 6))

# 1. Display Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# 2. Apply a Heatmap
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Normalize the grayscale image values to a range from 0 to 255
norm_gray_image = cv2.normalize(gray_image, None, 0, 255, cv2.NORM_MINMAX)

# Apply a color map to create a heatmap
heatmap_image = cv2.applyColorMap(norm_gray_image, cv2.COLORMAP_JET)

# Display heatmap side by side with the original image
plt.subplot(1, 2, 2)
plt.imshow(heatmap_image)
plt.title("Heatmap Overlay")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/Valley.jpg')  # Replace with the path to your image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for correct color display

# Create a figure with a 1x3 layout
plt.figure(figsize=(18, 6))

# 1. Display Original Image
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# 3. Increase Brightness
# Convert image to HSV (Hue, Saturation, Value)
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
hsv_image = np.array(hsv_image, dtype=np.float64)

# Increase the brightness by scaling the V (brightness) channel
brightness_scale = 1.5  # Adjust this value for more or less brightness
hsv_image[:, :, 2] = hsv_image[:, :, 2] * brightness_scale
hsv_image[:, :, 2][hsv_image[:, :, 2] > 255] = 255  # Cap the maximum at 255

# Convert back to RGB format
bright_image = cv2.cvtColor(np.array(hsv_image, dtype=np.uint8), cv2.COLOR_HSV2RGB)

# Display brightened image side by side with the original and heatmap images
plt.subplot(1, 3, 3)
plt.imshow(bright_image)
plt.title("Brightened Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/Valley.jpg')  # Replace with the path to your image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for correct color display

# Apply Canny Edge Detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Plot the original image and the edge-detected image side by side
plt.figure(figsize=(12, 6))

# Display Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Display Edge-Detected Image
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/Valley.jpg')  # Replace with the path to your image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for correct color display

# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)  # (15, 15) is the kernel size; adjust as needed

# Plot the original image and the Gaussian-blurred image side by side
plt.figure(figsize=(12, 6))

# Display Original Image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

# Display Gaussian-Blurred Image
plt.subplot(1, 2, 2)
plt.imshow(blurred_image)
plt.title("Gaussian Blur")
plt.axis("off")

plt.show()
```
